```python
# POST a Multiply-Encoded File

url = 'https://api.github.com/events'
files = {'file': open('Life_Expectancy_Data_v2.csv', 'rb')}

r = requests.post(url, files=files)
r.text
```

```python
r.status_code == requests.codes.ok
```
